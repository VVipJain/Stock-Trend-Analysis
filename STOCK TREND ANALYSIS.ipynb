STOCK ANALYSIS PROJECT


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df = pd.read_csv("NFLX.csv")
```

```python
sns.set(rc = {"figure.figsize" : (10,5)})
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index(df["Date"])
df
```

VOLUME OF STOCKS VS TIME

```python
sns.lineplot(x = df.index , y = df["Volume"], label = "Volume")
plt.title("VOLUME VS TIME")
```

```python
df.plot(y = ["High" , "Open" , "Close"] , title = "STOCK PRICE ANALYSIS")
```

VOLUME TRADED OVER DAY , MONTH AND YEAR

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize = (15,13))
df.groupby(df.index.day).mean().plot(y = "Volume" , ax = ax1 , xlabel = "Day")
df.groupby(df.index.month).mean().plot(y = "Volume" , ax = ax2 , xlabel = "Months")
df.groupby(df.index.year).mean().plot(y = "Volume" , ax = ax3 , xlabel = "Year")
```

DATES WITH HIGHEST STOCK PRICES

```python
a = df.sort_values(by = "High" , ascending = False).head(5)
a["High"]
```

DATES WITH LOWEST STOCK PRICES

```python
b = df.sort_values(by = "Low" , ascending = True).head(5)
b["Low"]
```

TRENDS OVER HIGHS AND LOWS

```python
fig, (ax1,ax2) = plt.subplots(nrows = 1 , ncols = 2 , sharex = True , figsize = (17,4))
fig.suptitle("HIGHS AND LOW TRENDS OVER TIME")
sns.lineplot(x = df.index , y = df["High"] , ax = ax1 , color = "green")
sns.lineplot(x = df.index , y = df["Low"] , ax = ax2 , color = "red")
```
